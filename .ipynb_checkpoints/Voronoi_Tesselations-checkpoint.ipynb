```python
import numpy as np
import pysal as ps
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib as mpl
import shapely
import time
from scipy.spatial import Voronoi, voronoi_plot_2d
%matplotlib inline

mpl.rcParams['figure.figsize'] = 100,100  #set the default map size
mpl.rcParams['patch.linewidth'] = 0.5  #set default polygon line width

np.random.seed(352)

# Local path on user's machine
path = '/Users/jgaboardi/Algorithms_etc./Data/'
```

```python
# Create Simulated Data for the Voronoi
simulated_points = 2000
simulated_range = range(simulated_points)
points_list = []
for i in simulated_range:
    points_list.append([float(np.random.uniform(0,1000,1)), float(np.random.uniform(0,1000,1))])

# Instantiate and Plot the Voronoi    
t1_simulated = time.time()
vor = Voronoi(points_list)
voronoi_plot_2d(vor)
plt.show()
t2_simulated = time.time()-t1_simulated
print t2_simulated, 'seconds to instantiate the Voronoi Tessellation and plot the diagram'
print vor.npoints, 'points'
print len(vor.regions), 'regions'
```

```python
simulated_lines = [shapely.geometry.LineString(vor.vertices[line])
                    for line in vor.ridge_vertices
                    if -1 not in line]
```

```python
simulated_gdf = gpd.GeoSeries([p for p in shapely.ops.polygonize(simulated_lines)])
```

```python
simulated_gdf.plot()
```

```python
simulated_gdf = gpd.GeoDataFrame(simulated_gdf)
```

```python
simulated_gdf.columns = ['geometry']
```

```python
# Waverly  Hills
streets = gpd.read_file(path+'Waverly_Trim/Waverly.shp')
streets.to_crs(epsg=2779, inplace=True) # NAD83(HARN) / Florida North
streets.to_file(path+'waverly/waverly.shp')
shp_waverly = ps.open(path+'waverly/waverly.shp')
```

```python
ntw = ps.Network(path+'waverly/waverly.shp')
shp_waverly = ps.open(path+'waverly/waverly.shp')
```

```python
intial_buffer = streets.buffer(50)  #Buffer
intial_buffer[:5]
```

```python
union_buffer = intial_buffer.unary_union  #Buffer Union
union_buffer = gpd.GeoSeries(union_buffer)
union_buffer.crs = streets.crs
final_buffer = gpd.GeoDataFrame(union_buffer, crs=streets.crs)
final_buffer.columns = ['geometry']
```

```python
np.random.seed(352)
x = np.random.uniform(shp_waverly.bbox[0], shp_waverly.bbox[2], 100)
np.random.seed(850)
y = np.random.uniform(shp_waverly.bbox[1], shp_waverly.bbox[3], 100)  
coords = zip(x,y)
coords = [shapely.geometry.Point(i) for i in coords]
sim_points = gpd.GeoDataFrame(coords)
sim_points.crs = streets.crs
sim_points.columns = ['geometry']
```

```python
intersection = [final_buffer['geometry'].intersection(p) for p in sim_points['geometry']]
intersection = gpd.GeoDataFrame(intersection, crs=streets.crs)
intersection.columns = ['geometry']
```

```python
# Add records that are points within the buffer
point_in = []
for p in intersection['geometry']:
    if type(p) == shapely.geometry.point.Point:
        point_in.append(p)
```

```python
points_gdf = gpd.GeoDataFrame(point_in, crs=streets.crs)
```

```python
points_gdf[:5]
```

```python
points_list = []
for i,j in points_gdf[0].iteritems():
    points_list.append([j.x, j.y])
print points_list 
```

```python
# Instantiate and Plot the Voronoi    
t1 = time.time()
vor = Voronoi(points_list)
voronoi_plot_2d(vor)
plt.show()
t2 = time.time()-t1
print t2, 'seconds to instantiate the Voronoi Tessellation and plot the diagram'
print vor.npoints, 'points'
print len(vor.regions), 'regions'
```

```python
lines = [
    shapely.geometry.LineString(vor.vertices[line])
    for line in vor.ridge_vertices
    if -1 not in line
]
```

```python
df = gpd.GeoSeries([p for p in shapely.ops.polygonize(lines)])
```

```python
voronoi_df = gpd.GeoDataFrame(df, crs=None)
voronoi_df.columns = ['geometry']
voronoi_df.to_file('Users/jgaboardi/Desktop/VORONOI.shp')
```
